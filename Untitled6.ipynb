```python
import numpy as np
import matplotlib.pyplot as plt
```

For Frequency 100 Hz

```python
t=np.linspace(0,100,1000)

y=np.sin((200*3.14)*t)

plt.figure(1)
plt.plot(t,y,"r",linewidth=2)
plt.xlabel("time(sec)")
plt.ylabel("Sin(wt)");
plt.legend("y(t)")

plt.grid()
```

```python
t=np.linspace(0,100,1000)

y1=np.sin((200*3.14)*t-2)
plt.figure(1)
plt.plot(t,y,t,y1,"r",linewidth=2)
plt.xlabel("time(sec)")
plt.ylabel("Sin(wt-2)");
plt.legend("y(t)")
plt.grid()
```

```python
t=np.linspace(0,100,1000)

y2=np.sin((200*3.14)*t+2)
plt.figure(1)
plt.plot(t,y,t,y2,"r",linewidth=2)
plt.xlabel("time(sec)")
plt.ylabel("Sin(wt+2)");
plt.legend("sin((200*3.14)*t),sin(((200*3.14)*t)+2)")
plt.grid()
```

```python
t=np.linspace(0,100,1000)

y3=np.sin((400*3.14)*t)
plt.figure(1)
plt.plot(t,y,t,y3,"r",linewidth=2)
plt.xlabel("time(sec)")
plt.ylabel("Sin(2wt)");
plt.legend("sin((200*3.14)*t),sin((400*3.14)*t)")
plt.grid()
```

# Continous Unit Step Function U(t)

```python
x=np.arange(-1,18,1)
y=np.arange(-1,8,1)
for i in list(y+1):
  if (i>=0):
    y[i]=1
  elif(i<0):
    y[i]=0


plt.plot(y)
plt.xlabel('time')
plt.ylabel('U(t)')
plt.title('Continous Unit Step Function')

```

# Discrete Unit Step Function U(t)

```python
x=np.arange(-1,8,1)
y=np.arange(-1,8,1)
for i in list(y+1):
  if (i>=0):
    y[i]=1
  elif(i<0):
    y[i]=0


plt.stem(y)
plt.xlabel('n Index')
plt.ylabel('U[n]')
plt.title('Discrete Unit Step Function')
```
